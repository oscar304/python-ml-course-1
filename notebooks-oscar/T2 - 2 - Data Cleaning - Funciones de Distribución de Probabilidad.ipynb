# Funciones de distribución de probabilidades

## Distribución uniforme

```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
a = 1
b = 100
n = 1000000
data = np.random.uniform(a,b,n)
```

```python
%matplotlib inline
plt.hist(data)
```

# Distribución Normal

```python
data = np.random.randn(1000000)
```

```python
x = range(1,1000001)
plt.plot(x, data)
```

```python
plt.hist(data)
```
